```python
import os
import razdel
from hazm import Normalizer, word_tokenize, stopwords_list
from rank_bm25 import BM25Okapi
import torch
from transformers import AutoTokenizer
from parsivar import BertForParsing

# Define the path to the directory containing the text files
directory_path = '/path/to/directory/'

# Read the documents from the directory
documents = []
for filename in os.listdir(directory_path):
    with open(os.path.join(directory_path, filename), 'r', encoding='utf-8') as f:
        text = f.read()
        documents.append(text)

# Preprocessing functions
normalizer = Normalizer()
stopwords = stopwords_list()

def preprocess(text):
    # Normalize the text
    normalized_text = normalizer.normalize(text)
    # Tokenize the text
    tokens = word_tokenize(normalized_text)
    # Remove stop words
    filtered_tokens = [token for token in tokens if token not in stopwords]
    # Return the preprocessed tokens
    return filtered_tokens

# Preprocess the documents
preprocessed_docs = [preprocess(doc) for doc in documents]

# Create a BM25 index based on the preprocessed documents
bm25 = BM25Okapi(preprocessed_docs)

# Define the query
query = "اپلیکیشن های ضروری برای سفر خارجی چیست؟"

# Preprocess the query
preprocessed_query = preprocess(query)

# Retrieve and rank the most relevant sentences
scores = bm25.get_scores(preprocessed_query)
sentence_scores = []
for i, doc in enumerate(preprocessed_docs):
    for j, sentence in enumerate(doc):
        score = scores[i]
        sentence_scores.append((i, j, score))

# Sort the sentence scores
sentence_scores = sorted(sentence_scores, key=lambda x: x[2], reverse=True)

# Define the number of sentences to include before and after the top-ranked sentences
num_context_sentences = 1

# Load the ParseBERT model and tokenizer
model_name = "HooshvareLab/bert-base-parsbert-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = BertForParsing.from_pretrained(model_name)

# Define the device to use for the model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Define a function to extract the most relevant noun phrases from a sentence
def extract_noun_phrases(sentence):
    # Tokenize the sentence using the ParseBERT tokenizer
    input_ids = torch.tensor([tokenizer.encode(sentence, add_special_tokens=True)], device=device)
    # Generate the parse tree using the ParseBERT model
    with torch.no_grad():
        outputs = model(input_ids)
        parse_tree = outputs.parse_tree[0]
    # Extract the noun phrases from the parse tree
    noun_phrases = []
    for node in parse_tree.traverse():
        if node.type == 'NP':
            noun_phrase = ' '.join([token.word for token in node.tokens])
            noun_phrases.append(noun_phrase)
    # Return the extracted noun phrases
    return noun_phrases

# Print the most relevant sentences with context and extracted noun phrases
for doc_index, sent_index, score in sentence_scores[:num_context_sentences]:
    sentences = razdel.sentenize(documents[doc_index])
    context_sentences = [sentences[i].text for i in range(max(0, sent_index-num_context_sentences), min(len(sentences), sent_index+num_context_sentences+1))]
    context = ' '.join(context_sentences)
    noun_phrases = extract_noun_phrases(context)
    print(f"Document {doc_index}, Sentence {sent_index}: {context.strip()}")
    print(f"Extracted Noun Phrases: {noun_phrases}\n")
```

```python
import re
from hazm import stopwords_list, Normalizer, stemmer, sent_tokenize

normalizer = Normalizer()
stop_words = set(stopwords_list())
stemmer = stemmer()

def preprocess_text(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()
    text = normalizer.normalize(text)  # normalize the text
    text = re.sub(r'[^\u0600-\u06FF\s]', '', text)  # remove non-Persian characters and punctuations
    sentences = sent_tokenize(text)
    tokenized_sentences = []
    for sentence in sentences:
        words = sentence.split()
        words = [stemmer.stem(w) for w in words if w not in stop_words]  # remove stop words and stem the words
        tokenized_sentences.append(' '.join(words))
    return tokenized_sentences
```
